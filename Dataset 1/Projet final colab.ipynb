```python
def Check_for_Credit_Scoring_Range(credit_score):

    if ( (credit_score == 0) or (credit_score <= 579) ):
        return "Very Poor"
    
    elif ( (credit_score == 580) or (credit_score <= 669) ):
        return "Fair"
    
    elif ( (credit_score == 670) or (credit_score <= 739) ):
        return "Good"

    elif ( (credit_score == 740) or (credit_score <= 779) ):
        return "Very Good"
    
    elif ( (credit_score == 800) or (credit_score <= 850) ):
        return "Excellent"

    else:
        return ""
df["Rate_Credit_Score"] = df["Credit_Score"].apply(lambda x: Check_for_Credit_Scoring_Range(x))
```

```python
#par tranche de 100k de revenu annuel, faire la moyenne du credit score
#puis l'integrer dans le credit score vide

def generate_group(revenue):
    quotient = revenue//100_000
    return quotient

df["Group"] = df["Annual_Income"].apply(lambda x: generate_group(x))

df_temp = df[["Group", "Credit_Score"]].groupby(by="Group").mean().reset_index()
df_temp = df_temp.rename(columns={"Credit_Score": "Mean Credit Score per Revenue Group"})
df = df.merge(df_temp, how="left", on="Group")
df["Credit_Score"] = df["Credit_Score"].fillna(df["Mean Credit Score per Revenue Group"])
df = df.drop(columns=["Mean Credit Score per Revenue Group","Group"])
```
