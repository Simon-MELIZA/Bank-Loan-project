```python
def Check_for_Age_Range(Cage):

    if (Cage < 30):
        return "inf à 30"
    
    elif ( (Cage >= 30) and (Cage < 40) ):
        return "de 30 à 40"
    
    elif ( (Cage >=40 ) and (Cage < 50) ):
        return "de 40 à 50"

    elif ( (Cage >=50) and (Cage < 60) ):
        return "de 50 à 60"
    
    elif (Cage >= 60):
        return "sup à 60"

    else:
        return ""
df["age"] = df["age"].apply(lambda Cage: Check_for_Age_Range(Cage))
```
